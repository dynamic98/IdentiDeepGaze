```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datapath = 'C:\\Users\\scilab\\IdentiGaze\\data\\DayDifference\\Similar_All.csv'
df = pd.read_csv(datapath, sep=',', header=0)
data = df.dropna(axis=0)

columnList = data.columns.tolist()
columnList.remove('participant')
columnList.remove('session')
```

```python
def getData(columnName):
    return data[['participant', 'session', columnName]]
def getAverage(data):
    return np.mean(data)
def getSSB(data):
    columnName = data.columns.tolist()[2]
    total_length = len(data[columnName])
    total_average = getAverage(data[columnName])
    participant_list = data['participant'].unique().tolist()
    sumofSquare = 0
    for participant in participant_list:
        participant_average = getAverage(data.loc[data['participant'] == participant, columnName])
        sumofSquare += (participant_average - total_average)**2
    ssb = sumofSquare*(total_length/len(participant_list))
    return ssb
def getSSJ(data):
    columnName = data.columns.tolist()[2]
    total_average = getAverage(data[columnName])
    total_length = len(data[columnName])
    session_list = data['session'].unique().tolist()
    sumofSquare = 0
    for session in session_list:
        session_average = getAverage(data.loc[data['session'] == session, columnName])
        sumofSquare += (session_average - total_average)**2
    ssj = sumofSquare*(total_length/len(session_list))
    return ssj
def getSS(data):
    columnName = data.columns.tolist()[2]
    total_average = getAverage(data[columnName])
    total_length = len(data[columnName])
    sumofSquare = 0
    for i in data[columnName].tolist():
        sumofSquare += (i - total_average)**2
    return sumofSquare

def calcICC(BMS, JMS, EMS, n, k):
    first = n*(BMS - EMS)
    second = n*BMS
    third = k*JMS
    fourth = (n*k - n - k)*EMS
    return first/(second + third + fourth)

def getICC(data):
    """
    return 4 values: BMS, JMS, EMS, ICC
    """
    columnName = data.columns.tolist()[2]
    participant_len = len(data['participant'].unique().tolist())
    session_len = len(data['session'].unique().tolist())
    ssb = getSSB(data)
    ssj = getSSJ(data)
    ss = getSS(data)

    ems = (ss - ssb - ssj)/((participant_len-1)*(session_len-1))
    bms = ssb/(participant_len-1)
    jms = ssj/(session_len-1)
    icc = calcICC(bms, jms, ems, participant_len, session_len)
    return bms, jms, ems, icc
```

```python
for columnName in columnList:
    df_specific = getData(columnName)
    bms,jms,ems,icc = getICC(df_specific)
    if icc > 0.1:
        print(columnName)
        print(icc)
```
